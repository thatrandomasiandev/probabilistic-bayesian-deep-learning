# Synthetic DGP Walkthrough

Validate heteroscedastic regression and binary classification generators with known ground truth.

```python
import sys
sys.path.insert(0, '../src')

import matplotlib.pyplot as plt
import numpy as np

from prob_ml.data.regression_dgp import RegressionDGPConfig, generate_regression_data
from prob_ml.data.classification_dgp import ClassificationDGPConfig, generate_classification_data
```

```python
reg = generate_regression_data(RegressionDGPConfig(n_samples=2000, heteroscedastic_strength=1.5, seed=42))
sigma = reg.ground_truth['sigma']
mu = reg.ground_truth['mu']

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].scatter(mu, reg.y, alpha=0.3, s=8)
axes[0].set_xlabel('mu(x)')
axes[0].set_ylabel('y')
axes[0].set_title('Heteroscedastic regression')
axes[1].hist(sigma, bins=40, edgecolor='k')
axes[1].set_title('Aleatoric noise sigma(x)')
plt.tight_layout()
plt.show()
```

```python
clf = generate_classification_data(ClassificationDGPConfig(n_samples=2000, label_noise=0.1, seed=42))
p_star = clf.ground_truth['p_star']

plt.hist(p_star, bins=40, edgecolor='k')
plt.title('Bayes-optimal probabilities p*(x)')
plt.xlabel('p*')
plt.show()
print('Positive rate:', clf.y.mean())
```
